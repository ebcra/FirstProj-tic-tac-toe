```python
def move_():
    #Placeholder för move
    move = ""
    # while loop som bara tillåter input mellan 1-9
    while move not in['1','2','3','4','5','6','7','8','9']:
        # Felmeddelande om inte 1-9
        move = input("Pick a number: 1-9 ")
    # När vi fått ett svar mellan 1-9 returnerar vi det värdet!
    return move
```
